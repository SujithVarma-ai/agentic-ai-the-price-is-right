## Now we need to set up a HuggingFace Token as a secret in Modal

1. Go to modal.com, sign in and go to your dashboard  
2. Click on Secrets in the nav bar  
3. Create new secret, click on Hugging Face, this new secret needs to be called **huggingface-secret** because that's how we refer to it in the code  
4. Fill in your key as HF_TOKEN and the value as your actual token hf_...  
5. Click done

## Go to llama.py

This code deploys the Llama 3.2 3B LLM on Modal using a T4 GPU. It loads the model from Hugging Face and generates text responses for user prompts.

```python
from llama import app, generate
```

We give a prompt (Never gonna..), and the Llama model predicts and generates the next words based on what it learned during training.

```python
with modal.enable_output():
    with app.run():
        result=generate.remote("Never gonna give you up, never gonna")
result
```

## Go to pricer_ephemeral.py

This code deploys a fine-tuned Llama 3.2 3B model on Modal using a T4 GPU. The model is optimized with 4-bit quantization and LoRA fine-tuning to predict product prices from textual descriptions.

This fine tuned model is from LoRA (open source)

```python
from pricer_ephemeral import app, price
```

Returns the predicted price

```python
with modal.enable_output():
    with app.run():
        result=price.remote("Apple iPhone 15 Pro Max, 256GB storage, titanium design, unlocked, excellent condition.")
result
```
